## Exploiting class inheritance in Python

```python
def poly_spi(t, side, angle):
    if side:
        t.forward(side)
        t.right(angle)
        poly_spi(t, side - 2, angle)
```

```python
with Terrarium(animate = True) as t:
    t1 = Turtle(t, 'coral')
    
    t1.point(1)
    poly_spi(t1, 100, 119)
```

```python
with Terrarium(animate = True) as t:
    t1 = Turtle(t, 'navy')
    
    t1.point(1)
    poly_spi(t1, 60, 65)
```

```python
with Terrarium() as t:
    t1 = Turtle(t)
    
    t1.left(20)
    for i in range(200):
        t1.forward(1)
```

```python
class ShakyTurtle(Turtle):
    def __init__(self, t, col = 'b', shakiness = 10):
        super().__init__(t, col)
        self.shakiness = shakiness
    
    def forward(self, distance):
        self.right(random.uniform(-1, 1) * self.shakiness)
        super().forward(random.uniform(0.75, 1.25) * distance)
```

```python
with Terrarium() as t:
    t1 = ShakyTurtle(t)
    
    t1.left(20)
    for i in range(200):
        t1.forward(1)
```

```python
with Terrarium(animate = True) as t:
    t1 = ShakyTurtle(t, 'navy')
    
    t1.point(1)
    poly_spi(t1, 60, 65)
```

```python
with Terrarium() as t:
    t1 = ShakyTurtle(t, 'navy', 5)
    t2 = ShakyTurtle(t, 'coral', 10)
    t3 = ShakyTurtle(t, 'seagreen', 50)
    
    t1.left(20)
    for i in range(200):
        for ti in [t1, t2, t3]:
            ti.forward(1)
```

```python
class PredatorTurtle(ShakyTurtle):
    def follow(self, prey, distance):
        angle = math.atan2((prey.pos[1] - self.pos[1]), (prey.pos[0] - self.pos[0])) / self.deg - self.angle
        self.left(angle)
        super().forward(distance)
```

```python
with Terrarium(animate = True) as t:
    t1 = ShakyTurtle(t, 'navy', 20)
    t2 = PredatorTurtle(t, 'coral', 10)
    
    for i in range(200):
        t1.forward(1)
        t2.follow(t1, 0.75)
        
    t1.point()
    t2.point()
```

```python
def move(turtle, pos):
    turtle.pos = pos
    
Turtle.move = move
```

```python
with Terrarium(animate = True) as t:
    t1 = ShakyTurtle(t, 'navy', 20)
    t2 = PredatorTurtle(t, 'coral', 10)
    
    t1.move((75, 75))
    
    for i in range(200):
        t1.forward(1)
        t2.follow(t1, 0.75)
        
    t1.point()
    t2.point()
```
