##### Functional programming
* main concern is managing flow of data between functions
* advantage #1: within limits, functional programming is formally provable; correctness of its result can be established by analyzing it
* also breaks problem into blocks; modular, composable, debuggable and testable code
* In Python, functions are first-class objects; like an object - passable to other functions, variable assignment, used in collections, etc.

## Making fractals with functional Python

```python
def snowflake(turtle,dist,level):
    for i in range(3):
        drawkoch(turtle,dist,level)
        turtle.right(120)

def drawkoch(turtle,distance,level):
    if level == 0:
        turtle.forward(distance)
    else:
        drawkoch(turtle,distance/3,level-1)
        turtle.left(60)
        drawkoch(turtle,distance/3,level-1)
        turtle.left(-120)
        drawkoch(turtle,distance/3,level-1)
        turtle.left(60)
        drawkoch(turtle,distance/3,level-1)
```

```python
with Terrarium() as t:
    drawkoch(Turtle(t), 100, 1)
```

```python
with Terrarium() as t:
    drawkoch(Turtle(t), 100, 2)
```

```python
with Terrarium() as t:
    drawkoch(Turtle(t), 100, 3)
```

```python
with Terrarium(animate = True) as t:
    drawkoch(Turtle(t), 100, 4)
```

```python
with Terrarium() as t:
    t1 = Turtle(t)
    
    def flake():
        snowflake(t1, 30, 3)
        
    def left():
        t1.left(60)
        
    for f in [flake, left] * 6:
        f()
```

### Function partials

```python
def over(a, b):
    return a/b
```

```python
inverse = functools.partial(over, 1)
half = functools.partial(over, b = 2)
```

```python
inverse(2)
```

```python
half(3)
```

```python
inverse
```

```python
def F(dist):
    return functools.partial(Turtle.forward, distance = dist)

def L(ang):
    return functools.partial(Turtle.left, angle = ang)
```

```python
f100 = F(100)

with Terrarium() as t:
    f100(Turtle(t))
```

```python
def func_koch(steps):
    new_steps = []
    
    for step in steps:
        if step.func is Turtle.forward:
            f3 = F(step.keywords['distance'] / 3)
            new_steps += [f3, L(60), f3, L(-120), f3, L(60), f3]
        else:
            new_steps += [step]

    return new_steps
```

```python
step0 = [F(100)]
```

```python
func_koch(step0)
```

```python
with Terrarium() as t:
    t1 = Turtle(t)
    
    for f in func_koch(func_koch(func_koch(step0))):
        f(t1)
```

```python
def iterated(f, n):
    def iter_f(arg):
        ret = arg
        for i in range(n):
            ret = f(ret)
        return ret
    
    return iter_f
```

```python
def turtle_map(self, steps):
    for step in steps:
        step(self)
        
Turtle.map = turtle_map
```

```python
with Terrarium() as t:
    Turtle(t).map(iterated(func_koch, 5) (step0))
```

```python
triangle = [F(100), L(-120), F(100), L(-120), F(100)]
```

```python
with Terrarium() as t:
    Turtle(t).map(iterated(func_koch, 5) (triangle))
```
