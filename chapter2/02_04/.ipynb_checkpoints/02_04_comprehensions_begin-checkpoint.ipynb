### Code and data needed from previous videos

List of olympic cities and years:

```python
cities = []
years = []

for game in open('games.txt','r'):
    words = game.split()
    
    city = ' '.join(words[:-1])
    year = words[-1].strip('()')

    cities.append(city)
    years.append(year)
```

Geolocated olympic-city coordinates (from a JSON file):

```python
coordinates_by_city = json.load(open('coords.json','r'))
coordinates_by_city
```

## Comprehensions and generators

__Comprehensions__ are a way to run a loop within a single line of code and to collect the results in a collection, such as a list.

__Generators__ are a shortcut to write functions that implement iterators.

```python
results = []

for city, year in zip(cities, years):
    if int(year) > 1945:
        results.append(f"{city}: {year}")
```

```python
results[:10]
```

#### List comprehension

```python
results = [f"{city}: {year}" for city, year in zip(cities, years) if int(year) > 1945]
```

```python
results[:10]
```

#### Dict comprehension

```python
cities_by_year = {year: city for city, year in zip(cities, years)}
```

```python
cities_by_year
```

#### Set comprehension
Useful if we only want unique elements

```python
cities_after_1930 = {city for year, city in cities_by_year.items() if int(year) > 1930}
```

```python
cities_after_1930
```

```python
pp.figure(figsize = (8, 8))

world = Basemap(projection = 'ortho', lat_0 = 75, lon_0 = 0)
world.drawcoastlines(linewidth = 0.25)
world.drawcountries(linewidth = 0.25)

for year, city in cities_by_year.items():
    # asterisk (*) - expands the tuple from coordinates_by_city() into two arguments for world()
    x, y = world(*coordinates_by_city[city])
    
    world.plot(x, y, 'r.')
    pp.text(x, y, year, fontsize = 8, ha = 'center', va = 'bottom', color = 'navy')
```

### Generator expressions

Useful if we do not need to store the comprehensions in a list and would like to consume them immediately.

Use *parentheses* instead of brackets or braces

```python
even = (i for i in range(20) if i % 2 == 0)
```

```python
even
```

```python
even.__next__()
```

```python
even.__next__()
```

```python
even = (i for i in range(20) if i % 2 == 0)
```

```python
sum(even)
```

```python
def fibonacci():
    f1, f2 = 0, 1
    
    while True:
        yield f2
        f1, f2 = f2, f1 + f2
```

1. Initial call of a generator suspends the execution and a generator object is returned

```python
f = fibonacci()
```

```python
f
```

2. When *\_\_next\_\_()* is called, execution resumes until the **yield** keyword which returns the value to the caller

```python
[next(f) for i in range(20)] # next(f) shorthand for f.__next__()
```
